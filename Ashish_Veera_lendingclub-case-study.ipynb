## Fixing the Missing Values
 -  "emp_length", "revol_util" are two columns having 1033 and 50 missing value respectively.

### Mode for Employee length is having higher frequency than the preceding length so mode can be used to impute missing value here. 

```python
print("Employee length Value Count -" ,loan_df.emp_length.value_counts())
print("mode of Employee length= " + loan_df.emp_length.mode()[0])

loan_df.emp_length.fillna(loan_df.emp_length.mode()[0], inplace = True)
print("Employee length Null Value Count -" , loan_df.emp_length.isna().sum())
```

### Dropping missing values in "revol_until"

```python
loan_df.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
print("Revol_Util Null Value Count -" , loan_df.revol_util.isna().sum())
```

### Data Standardization 
- "revol_util" & "int_rate" columns are object column, these have continous values with pecentage sign. 
- "emp_length" 10+ years can be considered as 10 while less than 1 can be considered as Zero '0' } 

```python
loan_df.emp_length = pd.to_numeric(loan_df.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
loan_df.revol_util = pd.to_numeric(loan_df.revol_util.apply(lambda x : x.split('%')[0]))
loan_df.int_rate = pd.to_numeric(loan_df.int_rate.apply(lambda x : x.split('%')[0]))
loan_df.nunique()
```
